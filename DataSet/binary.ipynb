```python
import pandas as pd
from fxpmath import Fxp
from sklearn.model_selection import train_test_split
from scipy import stats
import numpy as np
```

```python
def convert(x):
    y = Fxp(x, signed=True, dtype='S4.12')

    return int(y.bin(frac_dot=False), 2)
```

```python
# df = pd.read_csv('winequality-red.csv')
# df = pd.read_csv('insurance.csv')
df = pd.read_csv('airfoil_self_noise.dat', sep='\t')
df.head()
```

```python
df=(df-df.min())/(df.max()-df.min())
df[:10]
```

```python
df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
train, test = train_test_split(df, test_size=0.4, random_state=12)
```

```python
train.describe()
```

```python
test.describe()
```

```python
lines_train = []
for index, row in train.iterrows():
    x = []
    for j in row:
        x.append(str(hex(convert(j)))[2:])
    lines_train.append(' '.join(x))
```

```python
# Uncomment to write train file
with open('..\\sim_1\\new\\mem_large_5.mem', 'w') as f:
    f.write('\n'.join(lines_train))
```
